```python
import pandas as pd

df = pd.read_csv("df_ligas_completo.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
remove = [
    "fixture.id",              # ID da partida
    "fixture.referee",         # Nome do árbitro da partida
    "fixture.timezone",        # Fuso horário em que o jogo foi realizado (ex: UTC)
    "fixture.timestamp",       # Data/hora do jogo em formato numérico (Unix Timestamp)
    "fixture.periods.first",   # Timestamp exato do início do primeiro tempo
    "fixture.periods.second",  # Timestamp exato do início do segundo tempo
    "fixture.status.long",     # Descrição textual longa do status (ex: "Match Finished")
    "fixture.status.elapsed",  # Minutos de jogo decorridos (útil para jogos ao vivo)
    "fixture.status.extra",    # Informação extra do status (ex: tempo de prorrogação)
    "league.logo",             # URL para a imagem do logo da liga
    "league.flag",             # URL para a imagem da bandeira do país da liga
    "league.season",           # Ano da temporada da competição (ex: 2023)
    "league.round",            # Rodada ou fase da competição (ex: "Regular Season - 15")
    "league.standings",        # Dados aninhados da tabela de classificação da liga
    "teams.home.logo",         # URL para a imagem do logo do time da casa
    "teams.home.winner",       # Indicador de vitória (true/false) do time da casa, redundante com o placar
    "teams.away.logo",         # URL para a imagem do logo do time visitante
    "teams.away.winner",       # Indicador de vitória (true/false) do time visitante, redundante com o placar
    "score.fulltime.home",     # Gols do time da casa ao final do tempo regulamentar (90 min)
    "score.fulltime.away",     # Gols do time visitante ao final do tempo regulamentar (90 min)
    "score.halftime.home",     # Gols do time da casa no intervalo (1º tempo)
    "score.halftime.away",     # Gols do time visitante no intervalo (1º tempo)
    "score.extratime.home",    # Gols do time da casa na prorrogação
    "score.extratime.away",    # Gols do time visitante na prorrogação
    "score.penalty.home",      # Gols do time da casa na disputa de pênaltis
    "score.penalty.away",      # Gols do time visitante na disputa de pênaltis
]

df.drop(columns=remove, inplace=True)
```

```python
df.info()
```

```python
df["fixture.date"] = pd.to_datetime(df["fixture.date"]).dt.date
```

```python
import numpy as np

condicoes = [
    df["goals.home"] > df["goals.away"],
    df["goals.home"] < df["goals.away"],
    df["goals.home"] == df["goals.away"]
]

resultados = [
    "Vitória_Casa",
    "Vitória_Visitante",
    "Empate"
]

df["winner"] = np.select(condicoes, resultados, default=np.nan)
```

```python
df["winner.team"] = np.where(df["winner"] == "Vitória_Casa", df["teams.home.name"],
                             np.where(df["winner"] == "Vitória_Visitante", df["teams.away.name"], "-"))
```
